```python
#@title # (Optional) Mount google drive
#@markdown Mount google drive for result log file paths

from google.colab import drive
drive.mount('/content/drive')

DRIVE_PATH = "/content/drive/MyDrive/Whisper-Benchmark/results"
LOCAL_PATH = "/content/Whisper-WebUI-Benchmark/results"

import os
import shutil

os.makedirs(DRIVE_PATH, exist_ok=True)
os.makedirs(LOCAL_PATH, exist_ok=True)

if os.path.islink(LOCAL_PATH):
    os.remove(LOCAL_PATH)
    print(f"Removed existing symlink at {LOCAL_PATH}")
elif os.path.isdir(LOCAL_PATH):
    shutil.rmtree(LOCAL_PATH)
    print(f"Removed existing directory at {LOCAL_PATH}")

try:
    os.symlink(DRIVE_PATH, LOCAL_PATH)
    print(f"Successfully created symlink from {LOCAL_PATH} to {DRIVE_PATH}")
except OSError as e:
    print(f"Error occurred: {e}")
```
